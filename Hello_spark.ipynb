```
diamond_df = spark.read.format("csv")\
                .option("header","true")\
                .option("inferschema", "true") \
                .load("/databricks-datasets/Rdatasets/data-001/csv/ggplot2/diamonds.csv")

diamond_df.show(10)
```

```
from pyspark.sql.functions import avg

result_df = diamond_df.select("color", "price") \
                .groupBy("color")\
                .agg(avg("price")) \
                .sort("color")

result_df.show()
```
